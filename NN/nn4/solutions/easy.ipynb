```python
import os
import sys

sys.path.append(os.path.join('..'))
from models import *

%load_ext autoreload
%autoreload 2
```

```python
model_easy_softmax = MLP(
    data_dir=os.path.join('..', 'data'),
    architecture=MLPArchitecture(2, [20], 2),
    dataset_name="easy",
    last_layer_activation_func=Softmax(),
    loss_function="cross_entropy",
    target_precision=1e-6,
)
```

# Easy dataset - training and test data

```python
model_easy_softmax.plot_classification_data()
import matplotlib.pyplot as plt
```

## Training with Softmax activation on last layer
Using rmsprop with batches

```python
model_easy_softmax.train(
    epochs=1000,
    learning_rate=0.001,
    batch=True,
    optimizer="rmsprop",
    rms_beta=0.9,
    save_till_best=True
)
```

```python
model_easy_softmax.plot_history(smoothing_interval=1, scale="log")
```

## Predicted classification on test data
This dataset was simple enough for the model not to make any false predictions

```python
model_easy_softmax.plot_classification_comparison()
```

## Training with Sigmoid activation on last layer

```python
model_easy_sigmoid = MLP(
    data_dir=os.path.join('..', 'data'),
    architecture=MLPArchitecture(2, [20], 2),
    dataset_name="easy",
    last_layer_activation_func=Sigmoid(),
    loss_function="cross_entropy",
    target_precision=1e-6,
)
```

```python
model_easy_sigmoid.train(
    epochs=1000,
    learning_rate=0.001,
    batch=True,
    optimizer="rmsprop",
    rms_beta=0.9,
    save_till_best=True
)
```

```python
model_easy_sigmoid.plot_history(smoothing_interval=1, scale="log")
```
