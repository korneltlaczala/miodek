# Training with Sigmoid
Using a more complex architecture of `(2 - 20 - 20 - 3)`

```python
model_rings3_sigmoid = MLP(
    architecture=MLPArchitecture(2, [20, 20], 3),
    dataset_name="rings3-regular",
    data_dir="../data",
    last_layer_activation_func=Sigmoid(),
    loss_function="cross_entropy",
    target_precision=1e-6,
)

def train_sigmoid():
    model_rings3_sigmoid.train(
        epochs=1000,
        learning_rate=0.001,
        batch=True,
        optimizer="rmsprop",
        rms_beta=0.9,
        save_till_best=True
    )
```
