## Square simple

### Batch training

```python
arch = MLPArchitecture(1, [50, 100, 50], 1)
model_full = MLP(architecture=arch, dataset_name="square-simple")
model_full.evaluate()
model_full.plot()
model_full.train(epochs=3000, learning_rate=0.15)
model_full.train(epochs=3000, learning_rate=0.1)
model_full.train(epochs=3000, learning_rate=0.05)
model_full.evaluate()
model_full.plot()
```

### Mini-batch training

```python
arch = MLPArchitecture(1, [50, 100, 50], 1)
model_batch = MLP(architecture=arch, dataset_name="square-simple")
model_batch.evaluate()
model_batch.plot()
model_batch.train(epochs=1000, learning_rate=0.1, batch=True)
model_batch.train(epochs=1000, learning_rate=0.05, batch=True)
model_batch.train(epochs=1000, learning_rate=0.02, batch=True)
model_batch.evaluate()
model_batch.plot()
```

### Convergence rate

#### batch

```python
model_full.plot_history(start_age=100)
```

#### mini-batch

```python
model_batch.plot_history(start_age=100)
```

## Steps small

### Batch training

```python
arch = MLPArchitecture(1, [50, 100, 50], 1)
model_full = MLP(architecture=arch, dataset_name="steps-small")
model_full.evaluate()
model_full.plot()
model_full.train(epochs=3000, learning_rate=0.2)
model_full.train(epochs=40000, learning_rate=0.1)
model_full.train(epochs=40000, learning_rate=0.08)
model_full.train(epochs=20000, learning_rate=0.05)
model_full.evaluate()
model_full.plot()
```

### Mini-batch training

```python
arch = MLPArchitecture(1, [50, 100, 50], 1)
model_batch = MLP(architecture=arch, dataset_name="steps-small", name="mini-batch model")
model_batch.evaluate()
model_batch.plot()
model_batch.train(epochs=2000, learning_rate=0.2, batch=True)
model_batch.train(epochs=5000, learning_rate=0.1, batch=True)
model_batch.train(epochs=5000, learning_rate=0.08, batch=True)
model_batch.train(epochs=5000, learning_rate=0.05, batch=True)
model_batch.evaluate()
model_batch.plot()
```

### Convergence rate

#### batch

```python
model_full.plot_history(start_age=100)
```

#### mini-batch model

```python
model_batch.plot_history(start_age=100)
```
