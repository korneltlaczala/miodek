# Training with Sigmoid

```python
model_xor3_sigmoid = MLP(
    architecture=MLPArchitecture(2, [40, 40], 2),
    name="xor3_sigmoid",
    dataset_name="xor3",
    data_dir="../data",
    last_layer_activation_func=Sigmoid(),
    loss_function="cross_entropy",
    target_precision=1e-6,
)

def train_sigmoid(
            epochs=1000,
            learning_rate=0.001,
            batch=True,
            optimizer="rmsprop",
            rms_beta=0.9,
            save_till_best=True):
    model_xor3_sigmoid.train(
        epochs=epochs,
        learning_rate=learning_rate,
        batch=batch,
        optimizer=optimizer,
        rms_beta=rms_beta,
        save_till_best=save_till_best
    )
```

```python
train_sigmoid()
```

```python
model_xor3_sigmoid.plot_classification_comparison()
```

```python
train_sigmoid(learning_rate=0.001)
```

```python
model_xor3_sigmoid.train(
    epochs=1000,
    learning_rate=0.001,
    batch=True,
    batch_size=5,
    optimizer="rmsprop",
    rms_beta=0.9999,
    save_till_best=True
)
```

```python
model_xor3_sigmoid.train(
    epochs=1000,
    learning_rate=0.001,
    batch=True,
    batch_size=5,
    optimizer="rmsprop",
    rms_beta=0.9999,
    save_till_best=True
)
```

# Sigmoid activation
### Predicted classification on test data
For this dataset sigmoid activation function made it really difficult to achieve the target f1-score.

Final result achieved after extensive training was `0.94`. It is plausible, that the difficulty in improving the score was a result of overfitting, as loss on train set was orders of magnitude lower than loss on test set
